# Topic modeling hexameter corpus


```python
import sys;sys.path.append('..')
from ppanlp import *
pd.options.display.max_colwidth=None
```

```python
ppahex = PPASubcorpus('ppa_hexameter_corpus.csv')
```

## LDA Topic Modeling

```python
for ntopic in [10,25,50]:
    lda_topic_model = ppahex.topic_model(model_type='lda', min_doc_len=10)
    lda_topic_model.model(lemmatize=True,ntopic=ntopic,niter=1000,force=False)
    lda_topic_model.visualize_topics()
```

```python
lda_topic_model.visualize_topics()
```

## BERT Topic Modeling

```python
tm = ppahex.topic_model(min_doc_len=10)
tm.model(force=True)
```
